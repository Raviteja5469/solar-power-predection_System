### Univariate Analysis

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['generated_power_kw'], bins=30, kde=True)
plt.title('Generated Power Distribution')
plt.xlabel('Generated Power')
plt.ylabel('Frequency')
plt.show()
```

```python
df[df.columns[:9]].hist(figsize=(15, 15), bins=30)
plt.suptitle('Histograms of Weather Data')
plt.show()
```

```python
df[df.columns[9:18]].hist(figsize=(15, 15), bins=30)
plt.suptitle('Histograms of Weather Data')
plt.show()
```

```python
df[df.columns[18:-1]].hist(figsize=(15, 7), bins=30)
plt.suptitle('Histograms of Weather Data')
plt.show()
```

## Bivariate Analysis

```python
# Bivariate Analysis
plt.figure(figsize=(15, 30))
for i, column in enumerate(df.columns):
    plt.subplot(7, 3, i+1)
    sns.scatterplot(x=df[column], y=df['generated_power_kw'])
    plt.title(f'{column} vs Generated Power')
    plt.xlabel(column)
    plt.ylabel('Generated Power')
    
plt.tight_layout()
plt.show()
```

## Correlation Heatmap

```python

df.corr()
```

```python
plt.figure(figsize=(17, 12))
sns.heatmap(df.corr(),cmap='coolwarm' ,annot=True)
plt.show()
```

## Outlier Analysis

```python
# Outlier Detection
plt.figure(figsize=(15, 30))

for i, column in enumerate(df.columns):
    plt.subplot(7, 3, i+1)
    sns.boxplot(df[column])

plt.tight_layout()
plt.show()
```
